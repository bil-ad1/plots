```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pp
```

```python
x_p = np.array(["Eng","Math","Phy","Chem"])
y_p = np.array([45,65,55,75])
```

```python
df=pd.DataFrame(x_p,y_p,columns='marks')
print(df)
```

```python
pp.bar(x_p[0],y_p[0],color='r')
pp.bar(x_p[1],y_p[1],color='b')
pp.bar(x_p[2],y_p[2],color='g')
pp.bar(x_p[3],y_p[3],color='y')
pp.xlabel("Subjects")
pp.ylabel("Scores / 100")
pp.plot()
pp.show()
```
